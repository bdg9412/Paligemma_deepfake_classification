모델 로드

```python
from transformers import AutoProcessor, PaliGemmaForConditionalGeneration

from peft import PeftModel, PeftConfig
from transformers import AutoModelForCausalLM

# PeftConfig.from_pretrained 메서드를 사용하여 사전 학습된 PEFT 모델 설정을 로드
config = PeftConfig.from_pretrained("Donguri-b/paligemma_deepfake_2024")
base_model = PaliGemmaForConditionalGeneration.from_pretrained("google/paligemma-3b-pt-224")
# PeftModel.from_pretrained 메서드를 사용하여 기본 모델과 파인튜닝 된 모델의 PEFT 설정을 결합한 모델을 생성합니다.
model = PeftModel.from_pretrained(base_model, "Donguri-b/paligemma_deepfake_2024")
# 입력 데이터 처리를 위하여 AutoProcessor 로드
processor = AutoProcessor.from_pretrained("google/paligemma-3b-pt-224")
```

데이터셋 로드

```python
from datasets import load_dataset

ds = load_dataset("JamieWithofs/Deepfake-and-real-images")
```

```python
# 1(촬영된 사진) 위치 파악
for arg in range(len(ds['validation']['label'])):
    if ds['validation'][arg]['label'] == 1:
        print(arg)
        break
```

```python
# 학습에는 test 데이터를 활용하였으므로, validation 데이터로 검증 진행
real_image = ds['validation'][19641]['image']
fake_image = ds['validation'][0]['image']
```

```python
from PIL import Image
import requests


prompt = "Is this image made by AI?"
image_file = fake_image.convert("RGB")
```

```python
inputs = processor(prompt, image_file, return_tensors="pt")
output = model.generate(**inputs, max_new_tokens=20) # 결과 추론
```

```python
processor.decode(output[0], skip_special_tokens=True)[len(prompt)+1:] # 결과 확인
```
